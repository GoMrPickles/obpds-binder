```python
%matplotlib inline
from ipywidgets import interactive, fixed
from obpds import *

layers = [
    Layer(0.3*um, GaAs, -1e17/cm3), # n-type layer
    ]

d = TwoTerminalDevice(layers=layers,
                      contacts=[SchottkyContact(), OhmicContact()],
                      Fn='right')
```

```python
d.show_zero_current(V=0)
```

```python
interactive(d.show_zero_current, V=(-5,1), T=fixed(300), N=fixed(1000), approx=fixed('kane'))
```

```python
# print the simulated band profiles
r = d.get_zero_current(V=0)
print '{:<12}\t{:<12}\t{:<12}'.format('x (cm)', 'Ec (eV)', 'Ev (eV)')
for x, Ec, Ev in zip(r.x, r.Ec, r.Ev):
    print '{:.6e}\t{:.6e}\t{:.6e}'.format(x, Ec, Ev)
```
